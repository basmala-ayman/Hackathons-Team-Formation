```python
def to_list(col):
    if isinstance(col, list):
        return col
    if isinstance(col, str):
        try:
            return ast.literal_eval(col)
        except:
            return []
    return []
```

```python
def split_soft_from_hard(hard_skills):
    mapped_soft = []
    hard_remaining = []

    if not isinstance(hard_skills, list):
        return [], []

    for skill in hard_skills:
        s = str(skill).lower().strip()

        if s in SOFT_SKILL_MAP:
            mapped_soft.append(SOFT_SKILL_MAP[s])
        else:
            hard_remaining.append(skill)

    # remove duplicates (keep unique)
    mapped_soft = list(set(mapped_soft))
    hard_remaining = list(set(hard_remaining))

    return hard_remaining, mapped_soft
```

```python
members[["hard_skills_clean", "mapped_soft_skills"]] = members["hard_skills"].apply(
    lambda x: pd.Series(split_soft_from_hard(x))
)
```
