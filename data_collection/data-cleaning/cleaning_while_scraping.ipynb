```python
filtered_hacks.shape[0]
```

### Apply Filtering on Projects Table

```python
filtered_hacks_id = filtered_hacks["id"].unique()
filtered_hacks_id
```

```python
filtered_projects = projects[
    projects['Hackathon ID'].isin(filtered_hacks_id)
]
```

```python
filtered_projects.shape[0]
```

### Get Projects Count for each hackathon

```python
projects_per_hack = (
    filtered_projects.groupby('Hackathon ID')['Project Title'].count().reset_index(name = "Projects Count")
)
```

#### Add no. teams to its own hackathon (on hackathons table)

```python
filtered_hacks = filtered_hacks.merge(
    projects_per_hack,
    left_on = 'id',
    right_on = "Hackathon ID",
    how = 'left',
)
filtered_hacks.head()
```

```python
filtered_hacks = filtered_hacks.drop(columns='Hackathon ID')
filtered_hacks.head()
```

```python
filtered_hacks['Projects Count'] = filtered_hacks['Projects Count'].fillna(0).astype(int)
filtered_hacks.head()
```

```python
filtered_hacks = filtered_hacks[
    filtered_hacks["Projects Count"] > 5
]
```
